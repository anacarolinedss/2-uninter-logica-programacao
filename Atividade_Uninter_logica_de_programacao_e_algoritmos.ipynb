##**QUEST√ÉO 1 de 4 ‚Äì Conte√∫dos at√© Aula 3**
Enunciado: Imagina-se que voc√™ √© um dos programadores respons√°veis pela constru√ß√£o de um app de vendas para uma determinada empresa X que vende em atacado. Uma das estrat√©gias de vendas dessa empresa X √© dar desconto maior conforme o valor da compra, conforme a listagem abaixo:


‚Ä¢	Se valor for menor que 2500 o desconto ser√° de 0%;
‚Ä¢	Se valor for igual ou maior que 2500 e menor que 6000 o desconto ser√° de 4%;
‚Ä¢	Se valor for igual ou maior que 6000 e menor que 10000 o desconto ser√° de 7%;
‚Ä¢	Se valor for igual ou maior que 10000 o d250esconto ser√° de 11%;

Elabore um programa em Python que:

A.	Deve-se implementar o print com uma mensagem de boas-vindas que apare√ßa o seu nome e sobrenome [EXIG√äNCIA DE C√ìDIGO 1 de 6];
B.	Deve-se implementar o input do valor unit√°rio e da quantidade do produto [EXIG√äNCIA DE C√ìDIGO 2 de 6];
C.	Deve-se implementar o desconto conforme a enunciado acima (obs.: atente-se as condi√ß√µes de menor, igual e maior) [EXIG√äNCIA DE C√ìDIGO 3 de 6];
D.	Deve-se implementar o valor total sem desconto e o valor total com desconto [EXIG√äNCIA DE C√ìDIGO 4 de 6];

H.	Deve-se apresentar na sa√≠da de console um pedido recebendo desconto (valor total sem desconto maior ou igual a 2500) [EXIG√äNCIA DE SA√çDA DE CONSOLE 2 de 2];  





```python
#Inicio de boas vindas
print('Seja bem-vindo a loja da Ana Caroline')

#Dados
valor_unitario = float(input('Digite o valor unit√°rio do produto: '))
quantidade = int(input('Digite a quantidade do produto: '))
valor_total = valor_unitario * quantidade

#Conferindo desconto
if valor_total < 2500:
    desconto = 0
    print('Voc√™ n√£o tem desconto.')
elif valor_total >= 2500 and valor_total < 6000:
    desconto = 0.04
    print('Voc√™ ganhou desconto de 4%.')
elif valor_total >= 6000 and valor_total < 10000:
    desconto = 0.07
    print('Voc√™ ganhou desconto de 7%.')
else:
    desconto = 0.1
    print('Voc√™ ganhou desconto de 11%.')

#Calculando o valor com desconto
valor_com_desconto = valor_total - valor_total * desconto

#resultados
print('O valor SEM desconto √© de R$ {}'.format(valor_total))
print('O valor COM desconto √© de R$ {}'.format(valor_com_desconto))

#FIM DO CODIGO
```

```python
##INICIO
#Boas vindas AO CLIENTE
print('Seja bem-vindo a loja de Gelados da Ana Caroline')
print('-------------------Card√°pio------------------------')
print('---------------------------------------------------')
print('---|   Tamanho  |  Cupua√ßu (CP)  |  A√ßa√≠ (AC)  |---')
print('---|     P      |    R$  9,00    |  R$ 11,00   |---')
print('---|     M      |    R$ 14,00    |  R$ 16,00   |---')
print('---|     G      |    R$ 18,00    |  R$ 20,00   |---')
print('---------------------------------------------------')


#Siglas, sabores
sabores = {'CP': 'Cupua√ßu', 'cp': 'Cupua√ßu', 'AC': 'A√ßa√≠', 'ac': 'A√ßa√≠'}

#somando pedidos
soma = 0

while True:
#Solicitando sabor
    sabor = input('Entre com o sabor desejado (CP/AC): ')
    if sabor not in ['CP', 'AC', 'cp', 'ac']:
        print('Sabor inv√°lido. Tente novamente.')
        continue

#Solicita o tamanho
    tamanho = input('Entre com o tamanho desejado (P/M/G): ')
    if tamanho not in ['P', 'M', 'G', 'p', 'm', 'g']:
        print('Tamanho inv√°lido. Tente novamente.')
        continue

#Define o pre√ßo com base no sabor e no tamanho
    if sabor in ('CP', 'cp'):
        if tamanho in ('P', 'p'):
            preco = 9
        elif tamanho in ('M', 'm'):
            preco = 14
        else:
            preco = 18
    elif sabor in ('AC', 'ac'):
        if tamanho in ('P', 'p'):
            preco = 11
        elif tamanho in ('M', 'm'):
            preco = 16
        else:
            preco = 20

#Exibindo o pedido de acordo com o formato pedido pelo cliente
    if sabor in ('CP', 'cp'):
        print(f'Voc√™ pediu um Cupua√ßu no tamanho {tamanho.upper()}: R$ {preco:.2f}')
    elif sabor in ('AC', 'ac'):
        print(f'Voc√™ pediu um A√ßa√≠ no tamanho {tamanho.upper()}: R$ {preco:.2f}')

#acumulando valor do pedido
    soma += preco

#Pergunta caso o cliente pedir mais
    maisPedido = input('Deseja mais alguma coisa? (S/N): ')
    if maisPedido in ('N', 'n'):
        break

#Exibindo o valor total
print(f'O valor total a ser pago: R$ {soma:.2f}')

#FIM DO CODIGO
```
